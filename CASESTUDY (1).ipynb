# Case study on Presentation of data

Do the following operations on ‘cars_data’ dataset:
1. Read the dataset to the python environment.
2. Check for the null values present in the dataset.
3. Plot a bar graph of male vs female buyers participated in the sales.
4. Find the top 5 cars based on their sales price.
5. Find the least 5 cars based on their Resell price.

# 1. Read the dataset to the python environment.

```python
#importing_pandas_library
#importing_csv_file(‘cars_data’)

import pandas as pd

dat = pd.read_csv("cars_data.csv")
dat
```

# 2. Check for the null values present in the dataset.

```python
null = dat.isnull().values.any()

print("There are null values in the dataset[True/False] :",null)
```

# 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
#method_1_using_matplotlib

import matplotlib.pyplot as plt 

plt.figure(figsize=(8,8))
plt.hist(dat["Buyer Gender"],color='k')
plt.title('Graph of male vs female buyers participated in the sales',fontsize=15)
plt.xlabel('Buyer Gender')
plt.ylabel('Count')
```

```python
#method_2_using_seaborn

import seaborn as sns

sns.countplot(data = dat,x = 'Buyer Gender').set(title = "Graph of male vs female buyers participated in the sales")
```

# 4.Find the top 5 cars based on their sales price.

```python
#method_1

dat.sort_values('Sale Price',ascending=False).head(5)
```

```python
#method_2

dat.nlargest(5,'Sale Price')
```

# 5.Find the least 5 cars based on their Resell price.

```python
#method_1

dat.sort_values('Resell Price').head(5)
```

```python
#method_2

dat.nsmallest(5,'Resell Price')
```
